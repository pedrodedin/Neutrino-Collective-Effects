# Collective Effects - Bipolar System




## Collective Effects

Collective effects arise when we include neutrino-neutrino forward scattering interactions in the neutrino evolution. The Hamiltoninan/Potential for this process is given by [1,2]

$$    H_{\nu\nu,i} = \sqrt{2}G_F \sum_j  (1-\cos{\theta_{ij}}) (n_{\nu,j} \rho_{\nu,j} - n_{\overline{\nu},j} \rho_{\overline{\nu},j})\;\;\;\;\text{or} \;\;\;\;         H_{\nu\nu,\vec{p}} = \sqrt{2}G_F \int d^3p' (1-\hat{p}\cdot\hat{p}') (\rho_{\vec{p}'} - \overline{\rho}_{\vec{p}'} )$$

Therefore, the evolution equation becomes

$$ \newcommand{\ket}[1]{|#1\rangle} i \frac{d}{dt} \ket{\psi_{\nu,i}(t)} =  (\omega H_{V}+\lambda H_{\nu e}+\mu H_{\nu\nu, i}) \ket{\psi_{\nu,i}(t)}$$

where we extract the typical potential from each component of the Hamiltonian

$$\omega \equiv \frac{\Delta m^2}{2E_i},\;\;\;\; \lambda \equiv  \sqrt{2}G_F n_e ,\;\;\;\; \mu \equiv  \sqrt{2}G_F n_\nu$$

In the density matrix formalism

$$i \frac{d}{dt} \rho_i= [\omega H_{V}+ \lambda H_{\nu e}+\mu H_{\nu\nu,i}, \rho_i]$$

Decomposing in Pauli matrices

$$H_V = -\frac{1}{2}\vec{\sigma} \cdot \vec{B} \text{,} \;\;\;\;\;   H_{\nu e} = -\frac{1}{2}\vec{\sigma} \cdot \vec{L} \text{,} \;\;\;\;\; \rho= \frac{1}{2} \mathbf{1} + \frac{1}{2} \vec{\sigma} \cdot \vec{P}\text{,} \;\;\;\;\; \vec{\sigma}=(\sigma_1,\sigma_2,\sigma_3)$$

The evolution for the i-th neutrino and anti-neutrino polarization vectors ($\vec{P}_{\nu_i}$ and $\vec{P}_{\overline{\nu}_i}$) interacting with j (anti-)neutrinos become 

$$\frac{d}{dt}\vec{P}_{\nu_i} = \vec{P}_{\nu_i} \times \left[ \omega \vec{B} + \lambda \vec{L} + \mu \sum_j (1-\cos{\theta_{ij}}) (\frac{n_{\nu,j}}{n_\nu} \vec{P}_{\nu,j} - \frac{n_{\overline\nu,j}}{n_\nu}\vec{P}_{\overline{\nu},j}) \right]$$
    
$$\frac{d}{dt}\vec{P}_{\overline{\nu}_i} = \vec{P}_{\overline{\nu}_i} \times \left[ \omega \vec{B} - \lambda \vec{L}- \mu \sum_j (1-\cos{\theta_{ij}}) (\frac{n_{\nu,j}}{n_\nu} \vec{P}_{\nu,j} - \frac{n_{\overline\nu,j}}{n_\nu}\vec{P}_{\overline{\nu},j}) \right]$$

Note that if we want, we can define $\vec{P}$ to be normalized by $n_\nu$, we just need to be careful when traslating it back to the density matrix or to a conversion/survival probability. We can also turn this equation more compact using the neutrino falvor isospin (NFIS) notation $\vec{s}$ [3]

$$\frac{d}{dt}\vec{s}_{i} = \vec{s}_{i} \times \left[ \omega \vec{B} + \lambda \vec{L} + \mu \sum_j \mu_{ij} \vec{s}_{j}\right] = \vec{s}_{i} \times \left[ \omega \vec{B} + \lambda \vec{L} + \mu \vec{S} \right],\;\;\;\;  s_\nu \equiv \frac{\vec{P}_\nu}{2} \;\;\;\;s_{\overline{\nu}} \equiv - \frac{\vec{P}_{\overline{\nu}}}{2}$$

## Istoropic Monoenergetic Bipolar Neutrino Gas

* Istoropic gas $ \left \langle  \cos{\theta_{ij}}\right \rangle = 0$

* For simplicity, we will consider $\lambda = 0$. However, the matter effect can be easily introduced by changing the mixing values in vaccum to their correspondents in matter $\Delta m^2_V, \theta_V \rightarrow \Delta m^2_M, \theta_M$

* Bipolar systems are the ones formed by neutrinos with oposity isospin flavor (Ex: $\{\nu_e,\overline{\nu}_e\}$)

$$\frac{d}{dt}\vec{s}_{i} =  \vec{s}_{i} \times \left[ \omega_i \vec{B} + \mu \vec{S} \right]$$

* Starting with a monoenergetic symmetric bipolar system, i.e. equal amount/density $n_\nu$ of $\nu_e$ and $\overline{\nu}_e$, we will have two blocks os flavor isospi, one corresponding to neutrinos $\vec{S}_1(0) = +\hat{e}_3^F$ and other to the anti-neutrinos $\vec{S}_2(0) = -\hat{e}_3^F$ [3]

$$\frac{d}{dt}\vec{S}_{1}=  \vec{S}_{1}  \times \left[ \omega_1 \vec{B} + \mu( \vec{S}_1+\vec{S}_2) \right] =  \vec{S}_{1}  \times \left[ \omega_1 \vec{B} + \mu \vec{S}_2 \right]$$

$$\frac{d}{dt}\vec{S}_{2}=  \vec{S}_{2}  \times \left[ \omega_1 \vec{B} + \mu( \vec{S}_1+\vec{S}_2) \right]  =  \vec{S}_{2}  \times \left[ \omega_2 \vec{B} + \mu \vec{S}_1 \right]$$

This  sysmtem, as described analyticaly in [3,4] behaves like a pendulum that oscillates around the $\vec{B}$ vector for normal mass hierarchy (NH) and $-\vec{B}$ for inverted hierarchy (IH), due to the minus sign in $\omega_i=\Delta m^2/2E_i$.
```python
#Defining the set of funcitons that build up the ODE
def f(y, t, params):
    S1, S2, S3 = y      # unpack current values of y
    B1, B2, B3= params  # unpack parameters
    derivs = [S2*B3-S3*B2,      # list of dy/dt=f functions
              S3*B1-S1*B3,
              S1*B2-S2*B1]
    return derivs

def f_nu_antinu(y, t, params):
    nu1, nu2, nu3, antinu1, antinu2, antinu3 = y      # unpack current values of y
    B1, B2, B3= params  # unpack parameters
    derivs = [nu2*B3-nu3*B2,      # list of dy/dt=f functions
              nu3*B1-nu1*B3,
              nu1*B2-nu2*B1,
              -1*(antinu2*B3-antinu3*B2),      
              -1*(antinu3*B1-antinu1*B3),
              -1*(antinu1*B2-antinu2*B1)]              
    return derivs

def f_nu_antinu_coupled(y, t, params):
    nu1, nu2, nu3, antinu1, antinu2, antinu3 = y      # unpack current values of y
    B1, B2, B3, mu= params  # unpack parameters
    derivs = [nu2*B3-nu3*B2 +mu*(nu2*antinu3-nu3*antinu2) ,      # list of dy/dt=f functions
              nu3*B1-nu1*B3 +mu*(nu3*antinu1-nu1*antinu3) ,
              nu1*B2-nu2*B1 +mu*(nu1*antinu2-nu2*antinu1) ,
              -1*(antinu2*B3-antinu3*B2) +mu*(nu2*antinu3-nu3*antinu2),      
              -1*(antinu3*B1-antinu1*B3) +mu*(nu3*antinu1-nu1*antinu3),
              -1*(antinu1*B2-antinu2*B1) +mu*(nu1*antinu2-nu2*antinu1)]              
    return derivs

# Parameters
B= 2*math.pi/5 # B Magnitude prop to (\Delta m^2)/2E
theta=math.pi/10 #B angle -> depend on vaccum angle, and potential Acc(E,Ne)

B1 = - B*np.sin(theta)       
B2 = 0
B3 = B*np.cos(theta)      

# Initial values
S1_0 = 0   
S2_0 = 0     
S3_0 = 1.0      

# Parameters for ODE solver
params = np.array([B1, B2, B3])

# Initial conditions for ODE solver
#y0 = [S1_0, S2_0, S3_0]
y0=[0,0,1,0,0,1]

# Make time array for solution
tStop = 100.
tInc = 0.05
t = np.arange(0., tStop, tInc)

# Call the ODE solver
#psoln_nu = odeint(f, y0, t, args=(params,))
#psoln_antinu = odeint(f, y0, t, args=(params*(-1),))
#psoln_nu  = odeint(f_nu_antinu, y0, t, args=(params,))
psoln_nu_NH  = odeint(f_nu_antinu_coupled, y0, t, args=(np.array([B1, B2, B3,2]),))
psoln_nu_IH  = odeint(f_nu_antinu_coupled, y0, t, args=(np.array([-B1, -B2, -B3,2]),))

# Plot results
fig = plt.figure(1, figsize=(16,8))

ax1 = fig.add_subplot(221)
ax1.plot(t, 1/2*(1+psoln_nu_NH[:,2]),label='NH')
ax1.plot(t, 1/2*(1+psoln_nu_IH[:,2]),label='IH')
ax1.set_xlabel('time')
ax1.set_ylabel(r'$P_{\nu_e}$')
ax1.set_title('Neutrinos')
ax1.legend()

ax2 = fig.add_subplot(222)
ax2.set_title('Antineutrinos')
ax2.plot(t, 1/2*(1+psoln_nu_NH[:,5]),label='NH')
ax2.plot(t, 1/2*(1+psoln_nu_IH[:,5]),label='IH')
ax2.set_xlabel('time')
ax2.set_ylabel(r'$P_{\overline{\nu}_e}$')
ax2.legend()

ax3 = fig.add_subplot(223)
ax3.plot(t, 1/2*(1-psoln_nu_NH[:,2]),label='NH')
ax3.plot(t, 1/2*(1-psoln_nu_IH[:,2]),label='IH')
ax3.set_xlabel('time')
ax3.set_ylabel(r'$P_{\nu_\mu}$')
ax3.legend()

ax4 = fig.add_subplot(224)
ax4.plot(t, 1/2*(1-psoln_nu_NH[:,5]),label='NH')
ax4.plot(t, 1/2*(1-psoln_nu_IH[:,5]),label='IH')
ax4.set_xlabel('time')
ax4.set_ylabel(r'$P_{\overline{\nu}_\mu}$')
ax4.legend()

plt.tight_layout()
plt.show()
```

```python
G_F=1.1663787*10**(-23); #eV - Fermi Constant
delta_m2_31=2.5*10**(-3); #eV - \Delta m_31
theta_31=np.arcsin(math.sqrt(2.18*10**-2)); #\theta_31
N_A=6.02*10**(23) #Avogadro constant

#Calculate the effective mass squared difference in matter
def delta_m2_eff(delta_m2,theta,Acc):
  delta = math.sqrt((delta_m2*np.cos(2*theta)-Acc)**2+(delta_m2*np.sin(2*theta))**2)
  return delta

#Calculate the effective mixing angle in matter
def theta_eff(delta_m2,theta,Acc):
  theta_eff=(1/2)*math.atan2(1,1/((math.tan(2*theta))/(1-(Acc/(delta_m2*np.cos(2*theta))))))
  return theta_eff

#Calculate the matter "Potential" - Acc=2EV_cc
def Acc(N_e,E):
  A = 2*math.sqrt(2)*E*G_F*N_e
  return A

def B_vec(E):
  E=E*10**6 #From MeV to eV
  B=delta_m2_31/(2*E)
  B1=B*np.sin(2*theta_31)
  B2=0
  B3=B*np.cos(2*theta_31)
  return B,B1,B2,B3
```

```python
def f_nu_antinu_coupled_supernova(y, t, params):
    nu1, nu2, nu3, antinu1, antinu2, antinu3 = y      # unpack current values of y
    B1, B2, B3= params  # unpack parameters
    mu=mu_supernova(t)
    derivs = [nu2*B3-nu3*B2 +mu*(nu2*antinu3-nu3*antinu2) ,      # list of dy/dt=f functions
              nu3*B1-nu1*B3 +mu*(nu3*antinu1-nu1*antinu3) ,
              nu1*B2-nu2*B1 +mu*(nu1*antinu2-nu2*antinu1) ,
              -1*(antinu2*B3-antinu3*B2) +mu*(nu2*antinu3-nu3*antinu2),      
              -1*(antinu3*B1-antinu1*B3) +mu*(nu3*antinu1-nu1*antinu3),
              -1*(antinu1*B2-antinu2*B1) +mu*(nu1*antinu2-nu2*antinu1)]              
    return derivs
```
