## 5.1. Slow

```python
def func_Isotropic_Monoenergetic(time, y, params):
    omega,mu_opt,mu_0,lamb_opt,lamb_0,n_dim,theta= params  # unpack parameters
    
    B=np.array(B_vec(n_dim,theta))
    L=np.array(L_vec(n_dim))
    
    r=time/from_eV_to_1_over_km #From eV⁻¹ to km
    mu=mu_supernova(r,mu_opt,mu_0)
    lamb=lambda_supernova(r,lamb_opt,lamb_0)

    derivs=[]
    P_aux= cross_prod(y[0:n_dim],(B*omega+L*lamb+mu*y[n_dim:]))
    for k in range(n_dim):
        derivs.append(P_aux[k])
    
    #nu_bar 
    P_aux= cross_prod(y[n_dim:],(-1*B*omega+L*lamb-mu*y[0:n_dim]))
    for k in range(n_dim):
        derivs.append(P_aux[k])
        
    return derivs


def solver_Isotropic_Monoenergetic(P,E,r_i,r_f,mass_ord,mu_opt,mu_0,lamb_opt="no",lamb_0=0,n_f=2,theta=theta_31,delta_m2=delta_m2_31,solver="odeint"):

    omega=delta_m2/(2*E*10**6) #eV
    r_step = (2*np.pi/max(omega,mu_0))/200 #eV⁻¹
    r_i = r_i*from_eV_to_1_over_km
    r_f = r_f*from_eV_to_1_over_km
    r = np.arange(r_i,r_f,r_step) #eV⁻¹
    n_dim=(n_f**2)-1
    
    if mass_ord=="NH": 
        params=omega,mu_opt,mu_0,lamb_opt,lamb_0,n_dim,theta
    elif mass_ord=="IH":
        omega=-1*omega
        params=omega,mu_opt,mu_0,lamb_opt,lamb_0,n_dim,theta
    else:
        print("Not a mass ordering option!")
        return 0
        
    if solver=="odeint":
        y_sol= odeint(func_Isotropic_Monoenergetic, P, r, args=(params,),tfirst=True)
    else:
        psoln= solve_ivp(func_Isotropic_Monoenergetic,t_span=[r[0],r[-1]] ,y0=P, method=solver, args=(params,),t_eval=r,atol=1e-10,rtol=1e-7)
        y_sol=np.transpose(psoln.y)
        r=psoln.t
#     print(solver)
    psoln_trans=np.transpose(y_sol)
    
    P_nu=psoln_trans[0:3]
    P_nubar=psoln_trans[3:6]
    
    H_vac,H_nue=[],[]
    r=r/from_eV_to_1_over_km #From eV⁻¹ to km
    
    for r_i in r:
        B=np.array(B_vec(n_dim,theta_31))
        L=np.array(L_vec(n_dim))
        lamb=lambda_supernova(r,lamb_opt,lamb_0)
        H_vac.append(omega*B)
        H_nue.append(lamb*L)
        
    H_vac=np.transpose(H_vac)  
    H_nue=np.transpose(H_nue)   
    
    return P_nu,P_nubar,H_vac,H_nue,r
```

## Testing Bipolar Asymetric Case

```python
mu_opt ="const"
theta_V=10**-6 #rad
mu_0=10**5 #km⁻¹
mu_s=mu_0*10**(-3)*c_const #1/s
mu_0=mu_0/from_eV_to_1_over_km #eV
t_i,t_f=0,50000/mu_s #s
r_i,r_f=t_i,t_f*c_const*10**-3 #km
E_nu=10

# scenarios_slow=[[[0,0,1,0,0,0.8],E_nu,theta_V,"DOP853"],[[0,0,1,0,0,0.9],E_nu,theta_V,"DOP853"],[[0,0,1,0,0,0.99],E_nu,theta_V,"DOP853"]]
scenarios_slow=[[[0,0,1,0,0,0.9999],E_nu,theta_V,"DOP853"],[[0,0,1,0,0,0.995],E_nu,theta_V,"DOP853"],[[0,0,1,0,0,1],E_nu,theta_V,"DOP853"]]
```

```python
omega=delta_m2_31/(E_nu*10**6)
Q_vec=[0,0,2]-(omega/mu_0)*np.array(B_vec(3,theta_V))
D_vec=[0,0,1-0.995]
Q=np.sqrt(sum(Q_vec*Q_vec))
sigma=sum(Q_vec*D_vec)/Q
print("w= ",omega," eV")
print("u= ",mu_0," eV")
print("cos theta_max= ", mu_0*sigma**2/(2*omega*Q)-1)
print("sigma= ",sigma)
```

```python
#Slow
P_nu,t_slow=[],[]
for P, E, theta,solver in scenarios_slow:
    start_time = time.time()
    P_nu_aux,P_nubar_aux,H_vac_aux,H_nue_aux,r_aux=solver_Isotropic_Monoenergetic(P,E,r_i,r_f,"IH",mu_opt,mu_0,theta=theta,solver=solver)
    P_nu.append(P_nu_aux)
    t_slow.append(r_aux*10**3/c_const)
    print("IH took %s minutes" % ((time.time() - start_time)/60))
```

```python
plt.figure(figsize=(10,7))
for i in range(len(scenarios_slow)):
    plt.plot((10**6)*t_slow[i],1/2*(1-P_nu[i][2]/P_nu[i][2][0]),label=r"Slow Bipolar - $P_{\overline \nu}$= %.4f $P_{\nu}$"%(scenarios_slow[i][0][5]))
for i in range(len(scenarios_fast)):   
    plt.plot((10**6)*t_vec[i],P_ex_mean_time[i],label="Fast")
plt.xlabel("Time [$\mu$s]")
plt.ylabel(r"$\left \langle  P_{ex} \right \rangle_{\theta}$")
plt.legend()
plt.show()
```

### Slow

```python
theta_V=10**-6 #rad
mu_0=10**5 #km⁻¹
mu_s=mu_0*10**(-3)*c_const #1/s
mu_0=mu_0/from_eV_to_1_over_km #eV
t_i,t_f=0,10*10**(-6) #s
r_i,r_f=t_i,t_f*c_const*10**-3 #km
E_nu=10

# P=[0,0,1,0,0,1]
t_i,t_f=0,50000/mu_s #s
r_i,r_f=t_i,t_f*c_const*10**-3 #km

#Slow
# P=[0,0,1,0,0,1]
mu_opt ="const"

#Fast
theta_bins=100
theta_step=np.pi/theta_bins
cos_theta_step=2/theta_bins
nu_types=['nu_e','nu_x']
```

```python
#Slow
scenarios_slow=[[[0,0,1,0,0,1],E_nu,theta_V,"DOP853"],[[0,0,1,0,0,0.9],E_nu,theta_V,"DOP853"]]
P_nu,t_slow=[],[]
for P, E, theta,solver in scenarios_slow:
    start_time = time.time()
    P_nu_aux,P_nubar_aux,H_vac_aux,H_nue_aux,r_aux=solver_Isotropic_Monoenergetic(P,E,r_i,r_f,"IH",mu_opt,mu_0,theta=theta,solver=solver)
    P_nu.append(P_nu_aux)
    t_slow.append(r_aux*10**3/c_const)
    print("IH took %s minutes" % ((time.time() - start_time)/60))
```

### Fast

```python
scenarios_fast=[["NH","B",E_nu,"DOP853",0]]
```

```python
 nu,nubar=[],[]
t_vec=[]
for mass_ord,case,E_nu,solver,delta_m2 in scenarios_fast:
    start_time = time.time()
    theta_vec,t,mu_0, nu_aux, nubar_aux= solver_FFC_uniform(nu_types,t_i,t_f,E_nu,delta_m2,theta_V,mu_0,theta_bins,case,mass_ord,solver)
    print(solver,",",case," took %s minutes" % ((time.time() - start_time)/60))
    nu.append(nu_aux)
    nubar.append(nubar_aux)
    t_vec.append(t)
```

```python
rho_ex_theta_time=[]
rho_ex_mean_time=[]
P_ex_theta_time=[]
P_ex_mean_time=[]
for i in range(len(scenarios_fast)):
    rho_ex_theta_time_aux= 1/2*np.sqrt(np.transpose(nu[i][0][0])**2+np.transpose(nu[i][1][0])**2)
    rho_ex_theta_time.append(rho_ex_theta_time_aux)
    rho_ex_mean_time.append(sum(rho_ex_theta_time_aux)*cos_theta_step/2)
    
    P_ex_theta_time_aux=[]
    for j in range(theta_bins):
        P_ex_theta_time_aux.append(1/2*(1-np.transpose(nu[i][2][0])[j]/nu[i][2][0][0][j]))
    P_ex_theta_time.append(P_ex_theta_time_aux)
    P_ex_mean_time.append(sum(P_ex_theta_time_aux)*cos_theta_step/2)
```
